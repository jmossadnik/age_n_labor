#### First step analysis: Only keep core covariates and reduce frame of analysis in order to reduce data size

Furthermore, I excluded years before 1994 from the analysis, since the weight panlwt needed for the calculation of streams is only available for the years 1994 and following. (https://forum.ipums.org/t/weights-for-linking-cps-basic-monthly-data/2930)

Further more, I confine the analysis to individuals whose age ranges from 15 to 75 years.

```stata
keep year month cpsid cpsidp mish sex age empstat race panlwt wtfinl uhrs* wkstat
keep if inrange(age, 15, 75)
keep if year >= 1994
tab year
```

The original CPS data published by the BLS does not contain ids on an individual level, but on a household level **[CPSID]**.

Based on some characteristics that ought not to change across survey months, IPUMS created a individual id **[CPSIDP]**.

(Basically, the CPSIDP is initially constructed by the id CPSID of a household an individua appeared in the sample for the first time and her line number on the reporter's list of  interview partners of a household. Successive identifcation follows the simple idea that certain characteristics like **[RACE]** and **[SEX]** ought not change over time and **[AGE]** is allowed to change only in a certain interval. <br> (\\( age_{t+1} \in (age_t, age_t + 1) \forall \ \text{(MISH}_t, \text{MISH}_{t+1})  \neq (1, 8); \ \ age_{t+1} \in (age_t, age_t + 2) \ \text{for} \ \text{(MISH}_t, \text{MISH}_{t+1})  = (1, 8) \\), **[MISH]**: month in sample).\\(^1\\)




\\(^1\\) There is a 8 months break between MISH 4 and MISH 5, in which survey participants are not interviewed, allowing for the age to increase by max. 2 years from MISH 1 to MISH 8, which spans 12 months. This only matters if linking MISH 1 to MISH 8, for all consecutive months the age should increase by 1 year or not at all.

#### Check links from t to t-1

Some characteristics (sex, race, househould, age) should not change (that much, e.g. age) from month to month.

```stata
bysort cpsid cpsidp (mish): gen lag_mish = mish[_n - 1]
bysort cpsid cpsidp (mish): gen lag_sex = sex[_n - 1]
bysort cpsid cpsidp (mish): gen lag_age = age[_n - 1]
bysort cpsid cpsidp (mish): gen lag_race = race[_n - 1]
bysort cpsid cpsidp (mish): gen lag_year = year[_n - 1]
bysort cpsid cpsidp (mish): gen lag_month = month[_n - 1]
```

```stata
gen linked_b = 1 if sex == lag_sex & race == lag_race & inrange(year, lag_year, lag_year + 1) & ///
inrange(age, lag_age, lag_age + 2) & mish - lag_mish == 1 & ///
((mish != 5 & month == 1 & lag_month == 12) | (mish != 5 & month > 1 & month - lag_month == 1) | ///
(mish == 5 & lag_month - month == 3) | (mish == 5 & month - lag_month == 3))
```

```stata
tab linked_b, missing
```

#### Check links from t to t-2

```stata
bysort cpsid cpsidp (mish): gen lag2_mish = mish[_n - 2]
bysort cpsid cpsidp (mish): gen lag2_sex = sex[_n - 2]
bysort cpsid cpsidp (mish): gen lag2_age = age[_n - 2]
bysort cpsid cpsidp (mish): gen lag2_race = race[_n - 2]
bysort cpsid cpsidp (mish): gen lag2_year = year[_n - 2]
bysort cpsid cpsidp (mish): gen lag2_month = month[_n - 2]
```

```stata
gen linked_b2 = 1 if sex == lag2_sex & race == lag2_race & inrange(year, lag2_year, lag2_year + 1) & ///
inrange(age, lag2_age, lag2_age + 2) & mish - lag2_mish == 2 & ///
((mish != 5 & month == 1 & lag2_month == 11) | (mish != 5 & month == 2 & lag2_month == 12) | ///
(mish != 5 & month > 2 & month - lag2_month == 2) | ///
(mish == 5 & lag2_month - month == 4) | (mish == 5 & month - lag2_month == 4))
```

```stata
tab linked_b2, missing
```

#### Create recode for employment status

```stata
mvdecode empstat, mv(0)
recode empstat (1/12 = 1) (21 22 = 2) (32/max = 3), gen(emp)
label define emp 1 "E" 2 "U" 3 "N"
label value emp emp
tab emp
```

```stata
bysort cpsidp (mish): gen lag_emp = emp[_n - 1] if linked_b == 1
```

```stata
tab lag_emp, missing
```

```stata
recode lag_emp (. = 0)
label define lag_emp 0 "." 1 "E" 2 "U" 3 "N"
label value lag_emp lag_emp
tab lag_emp
```

#### Create employment flows from t-1 to t.

```stata
gen empflow = lag_emp * 10 + emp
label define empflow 01 ".E" 02 ".U" 03 ".N" 11 "EE" 12 "EU" 13 "EN" 21 "UE" 22 "UU" 23 "UN" 31 "NE" 32 "NU" 33 "NN"
label value empflow empflow
tab empflow
```

```stata
bysort cpsidp (mish): gen lag_panlwt = panlwt[_n - 1] if linked_b == 1 
```

```stata
bysort cpsidp (mish): gen lag_wtfinl = wtfinl[_n - 1] if linked_b == 1
```

```stata
gen flow_panlwt = (lag_panlwt + panlwt)/2
```

```stata
save "17nov2021", replace
```

```stata
egen flow_count = sum(flow_panlwt), by(year month empflow)
egen emp_count = sum(wtfinl), by(year month emp)
egen lag_emp_count = sum(lag_wtfinl), by(year month emp)
```

### create age group

```stata
** age group 1
recode age (15/19 = 1) (20/24 = 2) (24/29 = 3) (30/34 = 4) (35/39 = 5) (40/44 = 6) (45/49 = 7) (50/54 = 8) ///
(55/59 = 9) (60/64 = 10) (65/69 = 11) (70/74 = 12), gen(age_group1)

label define age_group1 1 "15-19" 2 "20-24" 3 "25-29" 4 "30-34" 5 "35-39" 6 "40-44" 7 "45-49" 8 "50-54" ////
9 "55-59" 10 "60-64" 11 "65-69" 12 "70-74"
label value age_group1 age_group1

** age group 2
recode age(15/29 = 1) (30/54 = 2) (55/74 = 3), gen(age_group2)

label define age_group2 1 "15-29" 2 "30-54" 3 "55-74"
label value age_group2 age_group2
```

```stata
gen date = ym(year,month)
format date %tmMCY
```

```stata
** counts for age group 1
egen flow_count_a1 = sum(flow_panlwt), by(year month empflow age_group1)
egen emp_count_a1 = sum(wtfinl), by(year month emp age_group1)
egen lag_emp_count_a1 = sum(lag_wtfinl), by(year month emp age_group1)
```

```stata
** counts for age_group 2
egen flow_count_a2 = sum(flow_panlwt), by(year month empflow age_group2)
egen emp_count_a2 = sum(wtfinl), by(year month emp age_group2)
egen lag_emp_count_a2 = sum(lag_wtfinl), by(year month emp age_group2)
```

```stata
gen employed_tag = 0
replace employed_tag = emp_count if emp == 1
egen employed_count = max(employed_tag), by(year month)
drop employed_tag

gen employed_tag = 0
replace employed_tag = emp_count_a1 if emp == 1
egen employed_count_a1 = max(employed_tag), by(year month age_group1)
drop employed_tag

gen employed_tag = 0
replace employed_tag = emp_count_a2 if emp == 1
egen employed_count_a2 = max(employed_tag), by(year month age_group2)
drop employed_tag
```

```stata
save "17nov2021", replace
```
